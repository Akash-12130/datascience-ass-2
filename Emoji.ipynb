```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import decomposition
```

```python
pca=decomposition.PCA(n_components=24,whiten=True , random_state=1)
```

```python
pca.fit(x_train)
```

```python
x_train_pca=pca.transform(x_train)
x_test_pca=pca.transform(x_test)
```

```python
x_train_pca.shape
```

```python
x_test_pca.shape
```

```python
eigen=(np.reshape(x[10],(350,350)).astype(np.float64))
```

```python
eigen
```

```python
fig=plt.figure(figsize=(30,30))
for i in range(10):
    ax=fig.add_subplot(3,5,i+1,xticks=[],yticks=[])
    ax.imshow(pca.components_[i].reshape(eigen.shape),cmap=plt.cm.bone)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClasssifier(n_estimators=100,random_state=42)
```

```python
model.fit(x_train_pca,y_train)
```

```python
y_pred=model.predict(x_test_pca)
```

```python
y_pred
```
